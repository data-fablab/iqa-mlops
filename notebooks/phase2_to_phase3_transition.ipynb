## 11. Repository checks

The following cells perform lightweight checks only.

They do not start Docker services, contact the IQA server, modify files, or execute production workflows.


````python
from pathlib import Path

def find_repo_root(start: Path | None = None) -> Path:
    """Find the IQA repository root from the current working directory."""
    current = (start or Path.cwd()).resolve()

    for candidate in [current, *current.parents]:
        if (candidate / "pyproject.toml").exists() and (candidate / "src").exists():
            return candidate

    return current


repo_root = find_repo_root()
print(f"Repository root: {repo_root}")
````

````python
from pathlib import Path

expected_paths = [
    "src/iqa",
    "tests",
    "docs",
    "deploy",
    "airflow/dags",
    "data/metadata",
    "configs",
    "scripts",
    "reports",
    ".github/workflows",
]

results = []

for relative_path in expected_paths:
    path = repo_root / relative_path
    results.append((relative_path, path.exists()))

width = max(len(name) for name, _ in results)

for name, exists in results:
    status = "OK" if exists else "MISSING"
    print(f"{name:<{width}}  {status}")
````

````python
import subprocess

def run_git_command(args: list[str]) -> str:
    """Run a read only Git command and return a readable result."""
    try:
        completed = subprocess.run(
            ["git", *args],
            cwd=repo_root,
            check=True,
            capture_output=True,
            text=True,
        )
        return completed.stdout.strip()
    except FileNotFoundError:
        return "Git is not installed."
    except subprocess.CalledProcessError as exc:
        message = exc.stderr.strip() or exc.stdout.strip()
        return f"Git command failed: {message}"


print("Branch:")
print(run_git_command(["branch", "--show-current"]) or "Detached HEAD")

print("\nCommit:")
print(run_git_command(["rev-parse", "--short", "HEAD"]))

print("\nWorking tree:")
status = run_git_command(["status", "--short"])
print(status or "Clean")
````

````python
from pathlib import Path

notebooks = sorted(
    path.relative_to(repo_root)
    for path in repo_root.rglob("*.ipynb")
    if ".git" not in path.parts and ".venv" not in path.parts
)

if notebooks:
    print("Notebooks found:")
    for notebook in notebooks:
        print(f"  {notebook}")
else:
    print("No notebooks found.")
````

````python
from pathlib import Path

kong_matches = sorted(
    path.relative_to(repo_root)
    for path in repo_root.rglob("*")
    if path.is_file()
    and ".git" not in path.parts
    and ".venv" not in path.parts
    and "__pycache__" not in path.parts
    and ".pytest_cache" not in path.parts
    and "kong" in path.name.lower()
)

if kong_matches:
    print("Kong related files found:")
    for path in kong_matches:
        print(f"  {path}")
else:
    print("No Kong related file was found in the current checkout.")
````
